#### Find latitude & longitude. Get url and json data

```python


address= 'Toronto, Ontario'
geolocator= Nominatim(user_agent='Foursquare_agent')
location=geolocator.geocode(address, timeout=10000)
latitude= location.latitude
longitude= location.longitude
print("latitude is : {}, longitude is : {}".format(latitude,longitude))
```

```python
#url 
radius=1000
new_url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, radius, LIMIT)
print(new_url)

#get url and json data
new_results=requests.get(url).json()
new_results
```

#### Get relevant part of JSON and transform it into a _pandas_ dataframe


```python
# assign relevant part of JSON to venues
venues = new_results['response']['groups'][0]['items']

# tranform venues into a dataframe
dataframe = json_normalize(venues)
dataframe.head()
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['venue.name', 'venue.categories'] + [col for col in dataframe.columns if col.startswith('venue.location.')] + ['venue.id']
dataframe_filtered = dataframe.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

dataframe_filtered
```

```python
df_part1= dataframe_filtered.copy()
df_part2= dataframe_filtered.copy() 
df_part3= dataframe_filtered.copy()
```

#### Get all the required data

```python
#Getting data for 1st part of assignment

df_part1= df_part1.drop('categories', axis=1)
df_part1= df_part1.drop('lat', axis=1)
df_part1= df_part1.drop('lng', axis=1)
df_part1= df_part1.drop('labeledLatLngs', axis=1)
df_part1= df_part1.drop('distance', axis=1)
df_part1= df_part1.drop('cc', axis=1)
df_part1= df_part1.drop('city', axis=1)
df_part1= df_part1.drop('state', axis=1)
df_part1= df_part1.drop('country', axis=1)
df_part1= df_part1.drop('formattedAddress', axis=1)
df_part1= df_part1.drop('address', axis=1)
df_part1= df_part1.drop('crossStreet', axis=1)
df_part1= df_part1.drop('id', axis=1)

#Extracting Postal code value
df_part1['postalCode']= df_part1.postalCode.str.split(expand=True)

#Renaming and rearranging columns
df_part1.rename(columns= {'name': 'Borough', 'postalCode': 'PostalCode', 'neighborhood' : 'Neighborhood'}, inplace=True)
df_part1=df_part1[['PostalCode', 'Borough', 'Neighborhood']]

#If a cell has a borough but a Not assigned  neighborhood, then the neighborhood will be the same as the borough
df_part1['Neighborhood'] = np.where(df_part1['Neighborhood'].isnull(), df_part1['Borough'], df_part1['Neighborhood']) 

df_part1.dropna(inplace=True)
df_part1
```

```python
#df_part1= df_part1.groupby('PostalCode')['Neighborhood'].apply(list)
df_part1= df_part1.groupby('PostalCode').agg(lambda x: list(set(x))).reset_index()
df_part1
```
